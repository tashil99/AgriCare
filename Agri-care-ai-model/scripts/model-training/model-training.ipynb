```python
from ultralytics import YOLO
import torch
import os

print("GPU Available:", torch.cuda.is_available())

DATA_YAML = "/content/drive/MyDrive/AgriCare-ai-model/dataset/data.yaml"


if not os.path.exists(DATA_YAML):
    raise FileNotFoundError(f"Dataset YAML not found at: {DATA_YAML}")

print("Dataset found!")


model = YOLO("yolov8s.pt")

# 8. Train model
model.train(
    data=DATA_YAML,
    epochs=50,
    imgsz=512,
    optimizer="AdamW",
    lr0=0.002,

    project="/content/drive/MyDrive/AgriCare-ai-model/trained-models",
    name="tomato_model_1",
)

# 9. Validate model
print("\nRunning validation...")
metrics = model.val()
print(metrics)

print("\nTraining Complete!")
```
